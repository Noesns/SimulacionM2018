# Sympy

```python
import sympy as sym
sym.init_printing(use_latex='mathjax')
```

```python
sym.var('x')
sym.sin(x)*np.sin(3)
```

```python
sym.sin(x)*sym.cos(x)*sym.tan(x)
```

## Funciones

```python
sym.var('x', real = True)
```

```python
f = x**2
f
```

```python
df = sym.diff(f,x)
```

```python
xc = sym.solve(df, x)
xc
```

```python
# convertir f e una función que se pueda evaluar numéricamente 
#(función lambdify de la librería sympy)
f_num = sym.lambdify([x], f, 'numpy')
x_vec = np.linspace(-5, 5, 100)

# graficar
plt.plot(x_vec, f_num(x_vec),color = "teal")
plt.xlabel('$x$',fontsize = 18)
plt.ylabel('$x^2$',fontsize = 18)
plt.title("Funciones", fontsize = 18)
plt.show()
```

## Como se definen funciones

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(x):
    return x**2
```

```python
def h(x,y):
    return np.sin(x)/np.tan(y)
```

```python
x = np.linspace(-5,5)
plt.plot(x,f(x),"--")
plt.plot(x,h(x,.1),"-.")
plt.scatter([0],[0], s=50)
plt.show()
```
